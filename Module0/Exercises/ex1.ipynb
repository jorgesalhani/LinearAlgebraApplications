## Exerc√≠cio 2

Retas $r$ e $l$ paralelas

```python
fig, ax = plt.subplots()
ax.plot(1,2, 'go', label='A')
ax.plot(3,-2, 'co', label='B')
ax.plot(-3,4, 'ro', label='C')
ax.plot(0,-2, 'mo', label='D')

t = np.linspace(-5,10, 100)
x_r = 1 + 2 * t
y_r = 2 - 4 * t

x_l = -3 + 3 * t
y_l = 4 - 6 * t


ax.plot(x_r, y_r, 'k-', label=r'Linha $r$')
ax.plot(x_l, y_l, 'b-', label=r'Linha $l$')
ax.legend(frameon=True)
plt.show()
```

Retas $r$ e $l$ perpendiculares

```python
fig, ax = plt.subplots()
ax.plot(1,2, 'go', label='A')
ax.plot(3,-2, 'co', label='B')
ax.plot(-3,4, 'ro', label='C')
ax.plot(0,11/2, 'mo', label='D')

t = np.linspace(-5,10, 100)
x_r = 1 + 2 * t
y_r = 2 - 4 * t

x_l = -3 + 3 * t
y_l = 4 + (3/2 * t)

ax.set_xlim(-10,8)
ax.set_ylim(-10,8)
ax.plot(x_r, y_r, 'k-', label=r'Linha $r$')
ax.plot(x_l, y_l, 'b-', label=r'Linha $l$')
ax.legend(frameon=True)
plt.show()
```

## Exerc√≠cio 4

```python
fig, ax = plt.subplots()

t = np.linspace(-5,10, 100)
y_r = 2 * t + 1
y_l = 3 * t - 2
y_s = 1 - t

ax.plot(3,7,'go', label='A')
ax.plot(0,1,'co', label='B')
ax.plot(3/4,1/4,'mo', label='C')

ax.set_xlim(-2, 4)
ax.set_ylim(-2, 10)

ax.plot(t, y_r, 'k-', label=r'Linha $r$')
ax.plot(t, y_l, 'b-', label=r'Linha $l$')
ax.plot(t, y_s, 'r-', label=r'Linha $s$')
ax.legend(frameon=True)
plt.show()
```
