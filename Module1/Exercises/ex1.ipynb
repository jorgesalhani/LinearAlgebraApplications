```python
import matplotlib.pyplot as plt
import numpy as np
import gc
```

## Exerc√≠cio 1

### (a)

```python
fig, ax = plt.subplots()
ax.plot(2, -1, 'bo', label='A')
ax.plot(-1,1, 'go', label='B')

t = np.linspace(-5, 10, 100)
x = -1 - 3 * t
y = 1 + 2 * t

ax.set_xlim(-3, 3)
ax.set_ylim(-3, 5)
ax.plot(x,y, 'k-', label=r'Linha $r$')
ax.legend(frameon = True, framealpha=1)
plt.show()
```

### (b)

```python
fig, ax = plt.subplots()
ax.plot(-5,5, 'go', label='A')

t = np.linspace(-5, 10, 100)
l = - t
x = -5 - 2 * t
y = 5 + 2 * t

ax.plot(t, l, 'b-', label=r'Linha $l$')
ax.plot(x, y, 'k-', label=r'Linha r')
ax.legend(frameon=True)
plt.show()
```

### (c)

```python
fig, ax = plt.subplots()
ax.plot(-4, -9, 'go', label='A')

t = np.linspace(-5, 10, 100)
x_l = 2 + 2 * t
y_l = 3 - t

x = -4 + 2 * t
y = -9 - t

ax.legend(frameon=True)
ax.plot(x_l, y_l, 'b-', label=r'Linha $l$')
ax.plot(x, y, 'k-', label=r'Linha $r$')
plt.show()
```
